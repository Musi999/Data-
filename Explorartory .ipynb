```python
# New data after cleaning
```

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')] 
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# Cleaning athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2.head()
```

```python
# Removing h from athlete perfomance 
```

```python
df2['Athlete performance']
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# Droping columns
```

```python
df2 = df2.drop(['Athlete club','Athlete year of birth','Athlete age category'], axis = 'columns')
```

```python
df2.head()
```

```python
# Null values
```

```python
df2.isna().sum()
```

```python
df2 = df2.dropna()
```

```python
 # reset index
```

```python
df2.reset_index(drop = True)
```

```python
# fix types
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2['Event dates'] = pd.to_datetime(df2['Event dates'], format='mixed')
```

```python
df2.head()
```

```python
df2.columns
```

```python
# rename columns
```

```python
df2 = df2.rename(columns = {'Year of event': 'Year', 
                           'Event dates' : 'race day',
                           'Event name' : 'race_name',
                           'Event distance/length' : 'race_length',
                           'Event number of finishers' : 'race_number_of_finishers',
                           'Athlete perfomance' : 'athlete_perfomance',
                           'Athlete gender' : 'athlete_gender',
                           'Athlete average speed' : 'athlete_average_speed',
                           'Athlete ID' : 'athlete_id'})
```

```python
df2.head()
```

```python
df3 = df2[['race day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','Athlete performance','athlete_average_speed','athlete_id']]
```
